```python
import pandas as pd
from IPython.display import display

df_pubmed = pd.read_csv('../data/treeoflife.species.pubmed.count.tsv', sep='\t', comment='#')
df_pubmed_ge1000 = df_pubmed[df_pubmed['Publication_count'] >= 1000].reset_index(drop=True)
display(df_pubmed_ge1000)
df_evolution = pd.read_csv('../data/treeoflife.species.evolution.tsv', sep='\t')
df_evolution = df_evolution[df_evolution['Species_ID'].isin(df_pubmed_ge1000['Species_ID'])].reset_index(drop=True)
display(df_evolution)
# find the row in df_evolution whose Species_ID is 9606
homo_sapiens_row = df_evolution[df_evolution['Species_ID'] == 9606]
display(homo_sapiens_row)
```
